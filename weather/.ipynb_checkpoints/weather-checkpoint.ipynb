```python
import pandas
import matplotlib.pyplot as plt
import datetime as dt
weather_file =pandas.read_csv('testset.csv')
```

# Clean data


```python
del weather_file[' _precipm']
del weather_file[' _hail']
del weather_file[' _snow']
del weather_file[' _tornado']
del weather_file[' _windchillm']

```

# Line Graphs

```python
#removing columns with String values from line graph

weather=weather_file
if ' _wdire' in weather.columns:
    del weather[' _wdire']
s= weather.columns
s=s[2:]
```

```python
fig ,axis = plt.subplots(nrows=3 , ncols=4 ,figsize=(16 , 12))
fig.subplots_adjust(hspace=0.5,wspace=0.5)
col =0
for j in range(0, len(s)):
    k=(j%4)
    data = weather_file[s[j]]
    x=[0]
    y=[0]
    for i in range(720,720*13, 720 ):
        if i==0:
            pass
        meandata = data[(i-720) : i ].mean()
        y.append(meandata)
        x.append((i/720)+1)
        
    axis[col][k].plot(x,y)
    axis[col][k].set_title(s[j])
    if k==3:
        col+=1    
```

# PIE Charts

```python
weather_file['Date'] = pandas.to_datetime(weather_file['datetime_utc'])
weather_file['month'] = weather_file['Date'].dt.month
g=weather_file.groupby('month')
```

```python
l={}
f=['January', 'February', 'March', 'April', 'May', 'June', 'July','August', 'September', 'October', 'November', 'December']
for i in range(1,13):
    t=g.get_group(i)
    count=t[' _conds'].value_counts()
    count=count[0:4]
    colors = ['r', 'm', 'g', 'b','b']
    #print(count)
    condition=count.index
    condition=condition[0:4]
    print(f[i-1])
    plt.pie(count, labels=condition, colors=colors, shadow=True,
       explode=(0,0,0.1,0), autopct="%1.1f%%", startangle=90)

    plt.legend(loc=(1,1))

    plt.show()
```
